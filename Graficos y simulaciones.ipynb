**a.** Generar, utilizando el comando rnorm(1000), 1000 elementos provenientes de una **normal estándar**, guarde los valores en el objeto x. Realizar un **histograma**, un **boxplot** y un **QQnorm** de x.

**Obtengo las mil muestras de una normal estándar y las coloco en un vector llamado x** 

```r
x <- rnorm(1000)
```

**Como ggplot solo puede trabajar con dataframes necesito convertir x a dataframe:**

```r
x <- data.frame(x)
```

**Dado que vamos a necesitar reescalar las imagenes creamos una función que hace esto:**

```r
reescalar <- function(ancho, alto){
 options(repr.plot.width = ancho, repr.plot.height = alto)
 }
```

```r
reescalar(ancho = 10, alto = 7)

ggplot(x, aes(x)) +
geom_histogram(bins = 30, col='black', fill='blue', alpha=0.4) + 
xlab('Valor observado') + 
ylab('Cantidad de observaciones') + 
ggtitle('Muestras de distribucion normal estandar - Histograma') + 
theme(title=element_text(size=20), axis.title=element_text(size=20), text = element_text(size=20)) 
```

```r
reescalar(ancho = 5, alto = 10)

ggplot(x, aes(x)) +
geom_boxplot(col='black', fill='blue', alpha=0.4) + 
xlab('Valor observado') + 
ggtitle('Normal estandar - Boxplot') + 
theme(title=element_text(size=20), axis.title=element_text(size=20), text = element_text(size=20)) +
coord_flip()
```

```r
reescalar(ancho = 7, alto = 7)

ggplot(x, aes(sample = x)) +
stat_qq() + 
ggtitle('Normal estandar - QQplot') + 
theme(title=element_text(size=20), axis.title=element_text(size=20), text = element_text(size=20))
```

**b.** Repetir el análisis anterior para las siguientes situaciones:

*iv.* **Gamma:** rgamma(1000,0.7)

```r
binomial <- data.frame(rbinom(1000, 10, 0.4))
chi <- data.frame(rchisq(1000, 50))
F <- data.frame(rf(1000,90,40))
gamma <- data.frame(rgamma(1000,0.7))
```

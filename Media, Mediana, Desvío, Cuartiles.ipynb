```r
library(ggplot2)
library(GGally)
```

```r
promedios <- read.table("graduados.txt")
head(promedios)
```

```r
colnames(promedios) <- c("Promedios")
promedios <- promedios$Promedios
```

**a.** Calcular la **media**, **la mediana muestral** y la **media 10 %-podada**.

```r
mean(promedios)
```

```r
median(promedios)
```

**b.** Calcular el **desvío estíandar**, **la distancia intercuartil** y la **MAD**.

```r
sd(promedios)
```

```r
IQR(promedios)
```

**c. Realizar un boxplot sobre este conjunto de datos. ¿Cuáles son las características más sobresalientes? ¿Cómo relaciona lo observado en el boxplot con lo obtenido en los incisos anteriores?**

```r
promedios <- data.frame(promedios)
```

```r
ggplot(promedios, aes(promedios)) +
geom_boxplot(col='black', fill='blue', alpha=0.4) + 
xlab('Promedios') + 
ggtitle('Promedios - Boxplot') + 
theme(title=element_text(size=20), axis.title=element_text(size=20), text = element_text(size=20)) +
coord_flip()
```

Una característica destacable es que la mediana es de 3.7. Esto indica que el 50% de los alumnos obtuvo un promedio de notas menor o igual a 3.7. Sin embargo en el sistema inglés esto es bueno ya que el **GPA** se mide entre 0 y 4.

* La **mediana** es la linea que separa las cajas del boxplot (3.7)
* El **rango intercuartílico** es la resta entre los limites de las cajas del boxplot (aproximádamente: 3.82-3.62 = 0.2)

**d. ¿Es razonable suponer normalidad de los datos? ¿Con qué gráfico o herramienta lo podría verificar?**

En general podremos suponer normalidad si los datos son "muchos", sin embargo esto es ambiguo. Un gráfico que nos permite visualizar la posibilidad de que exista normalidad es un histograma:

```r
ggplot(promedios, aes(promedios)) +
geom_histogram(col='black', fill='blue', alpha=0.4, bins = 13) + 
xlab('Promedio') + 
ylab('Cantidad de observaciones') + 
ggtitle('Promedios - Histograma') + 
theme(title=element_text(size=20), axis.title=element_text(size=20), text = element_text(size=20)) 
```

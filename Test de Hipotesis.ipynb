Consideraremos que una variable predictora $X_i$ **es significativa si no podemos rechazar la hipótesis nula** del siguiente test: 

$H_0$: $\beta_i$ = 0 $\hspace{10mm}$ vs $\hspace{10mm}$  $H_1$: $\beta_i$ $\neq$ 0

Observamos que $\beta_i$ es el coeficiente que multiplica a $X_i$ en el modelo de regresión lineal.

El test de hipótesis que se plantea para cada $\beta_i$ es el siguiente:

$\delta(X)$ =$\mathbb{1}${$ |T|> t_{n-p,\frac{\alpha}{2}} $}

Donde: 

- $T$ = $\frac{\hat{\beta_i}} {S\sqrt{d_{i,i}}}$

**Observación** el estadistico que usamos es:
$\frac{\hat{\beta_i}} {S\sqrt{d_{i,i}}}$ 
De manera que bajo $H_0$ verdadera (es decir $\beta_i= 0$),  $T$ distribuye t de student de n-p grados de libertad

- $d_{i,i}$ : elemento de la diagonal de ${(X^T X)}^{-1}$

- $\hat{\beta_i}$ = estimación del parametro $\beta_i$


De el resumen de la regresión realizada observamos: 

Llamando $\beta_1$ al parametro de la variable 'nox'

- $\hat{\beta_1}$ = -1.84 $\hspace{3mm}$ Es el coeficiente estimado para $\beta$

- $S\sqrt{d_{1,1}}$ = 2.65  $\hspace{3mm}$ es el 'Standard Error'

- t value = $\hat{\beta_1}$ / $(S\sqrt{d_{1,1}})$ = -0.69 $\hspace{3mm}$ (Es el valor del estadístico T para esta estimacion)
    
    
Ahora nos fijamos si este t value es mayor que el cuantil $t_{n-p, 1-\frac{\alpha}{2}}$ (n-p = 506 - 4) 

Tomando alfa = 0.05:


```r
n <- 506
p <- 4
alpha <- 0.05
qt(1-(alpha)/2, n-p)
```

Dado que mi **t-value es menor que el cuantil** digo que **no tengo suficiente información para rechazar la hipótesis nula** y por lo tanto considero que mi varible **nox es significativa**. **Observación:** esto no quiere decir que $\beta_1$ no sea 0. La conclusión fuerte de un test de hipótesis siempre resulta de rechazar la hipótesis nula. Es decir, si la rechazo puedo asegurar que es 0, pero si no la rechazo no puedo asegurar que no sea 0.
